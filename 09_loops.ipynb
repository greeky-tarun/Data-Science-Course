## Loops in Python

python has two main loops: for and while

## 1. For loop

Used to interate over sequences like lists, tuples and strings.

```python
fruits = ["apples", "banana", "cheery"]
for fruit in fruits:
    print(fruit)
```

### Using range()

```python
for i in range(3):
    print(i)
```

## 2. While loop

Runs as long as a condition is True.

```python
count = 0
while count < 3:
    print(count)
    count += 1
```

## 3. Loop Control Statements

- break -> Exits the loop
- continue -> Skips to the next iteration
- pass -> Does nothing[used as a placeholder]

```python
for i in range(5):
    if i == 3:
        break;
    print(i)
```

```python
l = [3, 5, 7, 78]
```

```python
for item in l:
    print(item)
```

```python
for i in range(1, 10):
    print(i)
```

```python
for i in range(10):
    if i == 0:
        continue;
    if i % 2 == 0:
        print(i)
```
