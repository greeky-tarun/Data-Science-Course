# String Formatting in Python

String is argubly the most used immutable data types in Python. Python provides multiple ways to format strings, including the format() method and f-strings(introduced in Python 3.6)

## 1. Using format()

```python
name = "Tarun"
age = 23
print("My name is {} and I am {} years old.".format(name, age))
```

### Positional and keyword Arguments

```python
print("{0} is learning {1}".format("Tarun","Python"))
```

```python
print("{name} is learning {language}".format(name = "Tarun", language = "Python"))
```

## 2. Using f-string (Recommended)

F-String provide a clearner and more readable way to format strings

```python
print(f"My name is {name} and I am {age} years old.")
```

### Expression inside f-strings

```python
a = 5
b = 10
print(f"Sum of {a} and {b} is {a+b}")
```

### Formatting Numbers

```python
pi = 3.24425
```

```python
print(f"Pi rounded to 2 decimal places: {pi:.2f}")
```

 ### Padding and Alignment


```python
print(f"{'Pyhton':<10}")
```

```python
print(f"{'Pyhton':>10}")
```

```python
print(f"{'Pyhton':^10}")
```

F-strings are the most efficinet and recommended way to format strings in modern Python
